```python
# Building the optimal model using backward elimination
import statsmodels.formula.api as sm
```

```python
startup_data.columns = ["Res", "Admin", "Market", "State", "Profit"]
```

```python
startup_data.head()
```

### Start 

```python
reg_model = sm.ols("Profit ~ Res + Admin+ Market + State", data=startup_data).fit()
```

```python
reg_model.summary()
```

```python
# lets remove state because it has highest p-value
reg_model_2 = sm.ols("Profit ~ Res + Admin+ Market", data=startup_data).fit()
```

```python
reg_model_2.summary()
```

```python
# lets remove "admin" because it has highest p-value
reg_model_3 = sm.ols("Profit ~ Res + Market", data=startup_data).fit()
```

```python
reg_model_3.summary()
```

```python
# lets remove "market" because it has highest p-value

reg_model_4 = sm.ols("Profit ~ Res", data=startup_data).fit()
```

```python
reg_model_4.summary()
```
